```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("stemlenmis_yorumlar.csv")

yorumlar = df["temizlenmis_cmleler"].astype(str).tolist()

kelimeler = []
for yorum in yorumlar:
    for kelime in yorum.lower().split():
        kelimeler.append(kelime)

kelime_sayac = {}
for kelime in kelimeler:
    kelime_sayac[kelime] = kelime_sayac.get(kelime, 0) + 1

sral_frekanslar = sorted(kelime_sayac.values(), reverse=True)

ranks = np.arange(1, len(sral_frekanslar) + 1)

plt.figure(figsize=(10, 6))
plt.loglog(ranks, sral_frekanslar, marker="o", linestyle="none", markersize=4, alpha=0.7, color="b")
plt.xlabel("Kelime Sras (log)")
plt.ylabel("Frekans (log)")
plt.title("Amazon Yorumlar in Zipf Yasas")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
