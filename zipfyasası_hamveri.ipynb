```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("amazon_yorumları.csv")

yorumlar = df["reviewText"].astype(str).tolist()

kelimeler = []
for yorum in yorumlar:
    for kelime in yorum.lower().split():
        kelimeler.append(kelime)

kelime_sayacı = {}
for kelime in kelimeler:
    kelime_sayacı[kelime] = kelime_sayacı.get(kelime, 0) + 1

sıralı_frekanslar = sorted(kelime_sayacı.values(), reverse=True)

ranks = np.arange(1, len(sıralı_frekanslar) + 1)

plt.figure(figsize=(10, 6))
plt.loglog(ranks, sıralı_frekanslar, marker="o", linestyle="none", markersize=4, alpha=0.7, color="b")
plt.xlabel("Kelime Sırası (log)")
plt.ylabel("Frekans (log)")
plt.title("Amazon Yorumları İçin Zipf Yasası")
plt.grid(True, which="both", linestyle="--", linewidth=0.5)
plt.show()
```
